```python
#import Packages
import librosa
```

**load audio file**

```python
audio_path="D:\Lab Setup\speech-to-text\AudioFiles\heart_beat_sound.wav"
x,sr=librosa.load(audio_path)
print(type(x),type(sr))
```

### Display Waveform

```python
# import packages

import matplotlib.pyplot as plt
import librosa.display
```

**Extract Zero Crossing Rate**

```python
plt.figure(figsize=(14,5))
librosa.display.waveshow(x, sr=sr)
```

### Zoom In and see the Variance in Deviation

```python
n0=9000
n1=9100
plt.figure(figsize=(14,5))
plt.plot(x[n0:n1])
plt.grid()
```
